```python
import CaboCha

c = CaboCha.Parser()

txt = input()

tree =  c.parse(txt)
cabocha_lattice = tree.toString(CaboCha.FORMAT_LATTICE)
```

```python
class Morph:
    def __init__(self,surface,attr):
        self.surface = surface
        self.base = attr[6]
        self.pos = attr[0]
        self.pos1 = attr[1]
```

```python
class Chunk:
    def __init__(self,morphs,dst,srcs):
        self.morphs = morphs
        self.dst = dst
        self.srcs = srcs
```

```python
from collections import defaultdict

res = []
for s in cabocha_lattice.split("EOS\n"):
    sent = []
    src_d = defaultdict(list)

    for line_ind, line in enumerate(s.split("\n")):

        if line == "":
            continue

        elif line[0] == "*":
            morphs = []

            _, ind, dst, *_ = line.split()
            dst = int(dst.replace("D", ""))
            src_d[dst].append(int(ind))

            if ind != 0:
                srcs = src_d[int(ind)]
                chunk = Chunk(morphs, dst, srcs)
                sent.append(chunk)
        else:
            surface, attr = line.split()
            attr = attr.split(",")
            morphs.append(Morph(surface, attr))

    res.append(sent)
```

```python
sent_lst = [("".join([mo.surface for mo in chunk.morphs if mo.pos != "記号"]),chunk.dst )for chunk in res[0]]
for i,j in sent_lst:
    print(i,sent_lst[j][0],sep="\t")
```

```python
from graphviz import Digraph

dg = Digraph(format='png')
for i,j in sent_lst:
    dg.node(i)
    
for i,j in sent_lst[:-1]:
    dg.edge(i,sent_lst[j][0])
```
