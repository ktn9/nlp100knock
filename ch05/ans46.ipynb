```python
class Morph:
    def __init__(self,surface,attr):
        self.surface = surface
        self.base = attr[6]
        self.pos = attr[0]
        self.pos1 = attr[1]
```

```python
class Chunk:
    def __init__(self,morphs,dst,srcs):
        self.morphs = morphs
        self.dst = dst
        self.srcs = srcs
```

```python
with open("ai.ja.txt.parsed",mode="r",encoding="utf-8") as f:
    txt = f.read()
txt = txt.replace("EOS\nEOS\n","EOS\n")
```

```python
from collections import defaultdict

res = []
for s in txt.split("EOS\n"):
    sent = []
    src_d = defaultdict(list)

    for line_ind, line in enumerate(s.split("\n")):

        if line == "":
            continue

        elif line[0] == "*":
            morphs = []

            _, ind, dst, *_ = line.split()
            dst = int(dst.replace("D", ""))
            src_d[dst].append(int(ind))

            if ind != 0:
                srcs = src_d[int(ind)]
                chunk = Chunk(morphs, dst, srcs)
                sent.append(chunk)
        else:
            surface, attr = line.split()
            attr = attr.split(",")
            morphs.append(Morph(surface, attr))

    res.append(sent)
```

```python
pred_case = []
for s in res:
    case_lst = []
    for chunk in s:
        case_lst.append([(morph.base,[morph.base for morph in chunk.morphs if morph.pos != "記号"]) for morph in chunk.morphs if morph.pos == "助詞"])
    for chunk in s:
        pred = ""
        case = []
        for mo in chunk.morphs:
            if "動詞" == mo.pos:
                pred = mo.base
                for ind in chunk.srcs:
                    for v in case_lst[ind]:
                        case.append(v)
                break
        if pred != "":
            case = sorted(list(case))
            pred_case.append((pred,case))
```

```python
for i,j in pred_case:
    tmp1 = []
    tmp2 = []
    for k,l in j:
        tmp1.append(k)
        tmp2.append("".join(l))
    print(i + "\t" + " ".join(tmp1) + "\t" + " ".join(tmp2))
```
