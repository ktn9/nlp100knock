```python
class Morph:
    def __init__(self,surface,attr):
        self.surface = surface
        self.base = attr[6]
        self.pos = attr[0]
        self.pos1 = attr[1]
```

```python
with open("ai.ja.txt.parsed",mode="r",encoding="utf-8") as f:
    txt = f.read()
txt = txt.replace("EOS\nEOS\n","EOS\n")
```

```python
res = []
for s in txt.split("EOS\n"):
    sent = []
    for line in s.split("\n"):
        if line == "":
            continue
        if line[0] != "*" and line != "EOS\n":
            surface,attr = line.split()
            attr = attr.split(",")
            sent.append(Morph(surface,attr))
    res.append(sent)
```

```python
for i in res[1]:
    print(vars(i))
```
